```python
import random
import numpy
def rueda1():
    x=random.random()
    if x < 2/22:
        resu=1 #"Cerezas"
    elif x < 4/22:
        resu=2 #"Naranjas"
    elif x < 9/22:
        resu=3 #"Ciruelas"
    elif x < 19/22:
        resu=4 #"Campanas"
    elif x < 21/22:
        resu=5 #"Barras"
    else:
        resu=6 #"7"
    return resu
    
def rueda2():
    x=random.random()
    if x < 5/22:
        resu=1 #"Cerezas"
    elif x < 8/22:
        resu=2 #"Naranjas"
    elif x < 9/22:
        resu=3 #"Ciruelas"
    elif x < 11/22:
        resu=4 #"Campanas"
    elif x < 21/22:
        resu=5 #"Barras"
    else:
        resu=6 #"7"
    return resu

def rueda3():
    x=random.random()
    if x < 2/22:
        resu=1 #"Cerezas"
    elif x < 9/22:
        resu=2 #"Naranjas"
    elif x < 19/22:
        resu=3 #"Ciruelas"
    elif x < 20/22:
        resu=4 #"Campanas"
    elif x < 21/22:
        resu=5 #"Barras"
    else:
        resu=6 #"7" 
    return resu      

J=1000
res=[]
for i in range(J):
  res+=[0]
  x=[rueda1(),rueda2(),rueda3()]
  if x==[6,6,6]:
    res[i]+=100
  elif x==[5,5,5]:
    res[i]+=50
  elif x==[4,4,4]:
    res[i]+=20
  elif x==[3,3,3]:
    res[i]+=20
  elif x==[2,2,2]:
    res[i]+=20
  elif x==[1,1,1]:
    res[i]+=20
  elif x[0]==1:
    res[i]+=2
  elif x[2]==1:
    res[i]+=2
  elif x.count(1)==2:
    res[i]+=5
  elif x==[2,2,5]:
    res[i]+=10
  elif x==[5,2,2]:
    res[i]+=10
  elif x==[3,3,5]:
    res[i]+=14
  elif x==[5,3,3]:
    res[i]+=14
  elif x==[4,4,5]:
    res[i]+=18
  elif x==[5,4,4]:
    res[i]+=18
t=(sum(res)-J)

if t > J:
  print("En",J, "juegos su ganancia es de: " + str(abs(t)))
elif t == J:
  print("En",J, "juegos no tuvo pérdida ni ganancia")
else:
  print("En",J, "juegos su pérdida es de: " + str(abs(t)))

media=(t-J)/J
desviacion=0
for x in res:
  desviacion=desviacion+(x - media)**2
desviacion= numpy.sqrt((desviacion / (J-1)))
print("La media es:",media)
print("La desviacion estandar es:",desviacion)

h=(1.96*(desviacion/(numpy.sqrt(J))))
inter_sup= (media + h)
inter_inf= (media - h)
print("La oscilación en",J,"juegos es:",round(h,5))
print("El intervalo de confianza al 95% para la ganacia o pérdida total es:",round(inter_inf,5),"--",round(inter_sup,5))

print("")

J=10000
res=[]
for i in range(J):
  res+=[0]
  x=[rueda1(),rueda2(),rueda3()]
  if x==[6,6,6]:
    res[i]+=100
  elif x==[5,5,5]:
    res[i]+=50
  elif x==[4,4,4]:
    res[i]+=20
  elif x==[3,3,3]:
    res[i]+=20
  elif x==[2,2,2]:
    res[i]+=20
  elif x==[1,1,1]:
    res[i]+=20
  elif x[0]==1:
    res[i]+=2
  elif x[2]==1:
    res[i]+=2
  elif x.count(1)==2:
    res[i]+=5
  elif x==[2,2,5]:
    res[i]+=10
  elif x==[5,2,2]:
    res[i]+=10
  elif x==[3,3,5]:
    res[i]+=14
  elif x==[5,3,3]:
    res[i]+=14
  elif x==[4,4,5]:
    res[i]+=18
  elif x==[5,4,4]:
    res[i]+=18
t=(sum(res)-J)

if t > J:
  print("En",J, "juegos su ganancia es de: " + str(abs(t)))
elif t == J:
  print("En",J, "juegos no tuvo pérdida ni ganancia")
else:
  print("En",J, "juegos su pérdida es de: " + str(abs(t)))

media=(t-J)/J
desviacion=0
for x in res:
  desviacion=desviacion+(x - media)**2
desviacion= numpy.sqrt((desviacion / (J-1)))
print("La media es:",media)
print("La desviacion estandar es:",desviacion)

h=(1.96*(desviacion/(numpy.sqrt(J))))
inter_sup= (media + h)
inter_inf= (media - h)
print("La oscilación en",J,"juegos es:",round(h,5))
print("El intervalo de confianza al 95% para la ganacia o pérdida total es:",round(inter_inf,5),"--",round(inter_sup,5))
```
